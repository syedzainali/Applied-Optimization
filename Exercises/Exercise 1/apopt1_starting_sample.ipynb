```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions, derivatives are only interesting for demonstration
def g(x,y):
    return (x-1)**2+100*(x**2-y)**2


def f(x, y):
    return 0.01*(x**2+y**2)-0.5*(np.cos(3*x-3*y)+np.cos(3*x+3*y))

# plot heatmap of function
def heatmap(func,ax,ran=(-3,3),delta=0.025):

    x = y = np.arange(ran[0],ran[1], delta)
    X,Y = np.meshgrid(x,y) # Take advantage of implicit numpy vectorization - no need for a loop here

    Z = func(X,Y)

    im = ax.imshow(Z, interpolation='bilinear', cmap="inferno",
                origin='lower', extent=[ran[0],ran[1],ran[0],ran[1]],
                vmax=abs(Z).max(), vmin=-abs(Z).max())
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,2,dpi=225)
heatmap(f,ax=axs[0],ran=(-10,10))
heatmap(g,ax=axs[1],ran=(-3,3))

# Do something ...

# ... scatter something
axs[0].scatter(2,0,color="black")
axs[1].scatter(2,2,color="black")
```
